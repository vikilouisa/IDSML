```python
#INTRODUCTION TO PANDAS
#https://campus.datacamp.com/courses/data-manipulation-with-pandas/transforming-dataframes?ex=1

#Pandas are a data manipulation library in Python.

#DataFrames
#Panda is based on NumPy and matplotlib.
#print(....heads()) to print first 5 rows of dataframe to quickly explore the data
#print(....info()) to get a concise summary of the dataframe snd if there are missing values
#print(....shape) to get the number of rows and columns in the dataframe
#print(....describe()) to get statistical summary of numerical columns in the dataframe
#print(....values) to get the underlying numpy array of the dataframe
#print(....columns) to get the column names of the dataframe


#IMPORTING DATA AS PANDA DATAFRAMES
#https://www.youtube.com/watch?v=dUpyC40cF6Q&list=PLUaB-1hjhk8FE_XZ87vPPSfHqb6OcM0cF&index=59

#!!!Reading in CSV-Files!!!
#import pandas as pd
#pd.read_csv("file.csv") to read in CSV files
#copy as path to get the file path of the file you want to read in --> it mostly be need to be read as raw so: pd.read_csv(r"file_path")
#shift + tab to get the docstring of a function in jupyter notebook
#if there is no header in the file you are reading in, use header=None to avoid the first row being used as column names: pd.read_csv("file.csv", header=None)
#header can be renamend by using names parameter: pd.read_csv("file.csv", names=["col1", "col2", "col3"])
#to set a specific column as the index while reading in the file, use index_col parameter: pd.read_csv("file.csv", index_col="col_name")
#df = pd.read_csv("file.csv") to store the dataframe in a variable called df 

#!!!Reading in txt-Files!!!
#df = pd.read_csv("file.txt") to read in txt files
#maybe a seperater is needed --> use sep parameter: pd.read_csv("file.txt", sep="\t") for tab seperated files
#this is improperly ---> BETTER: df = pd.read_table("file.txt") to read in txt files, no need to specify sep parameter for tab seperated files

#!!!Reading in json-Files!!!
#df = pd.read_json("file.json") to read in json files

#!!!Reading in excel-Files!!!
#df = pd.read_excel("file.xlsx") to read in excel files
#if you want to read in a specific sheet, use sheet_name parameter: pd.read_excel("file.xlsx", sheet_name="Sheet1")

#!!!Showing maximum number of columns and rows!!!
#pd.set_option("display.max_columns", None) to show all columns
#pd.set_option("display.max_rows", None) to show all rows
#df2.info() to check the info of the dataframe df, this shows the number of non-null values in each column (non-null values = values that are not missing)
#df2.shape to get the number of rows and columns in the dataframe df2
#df2.head(5) to print the first 5 rows of the dataframe df2
#df2.tail(3) to print the last 3 rows of the dataframe df2
#df2.describe() to get the statistical summary of the numerical columns in the dataframe df2
#df2.loc[:, "column_name"] to access a specific column in the dataframe df2 by its name
#df2.iloc[:, 0] to access a specific column in the dataframe df2 by its index (0 = first column)
#df2['Rank'] to access the column "Rank" in the dataframe df2

#INTRODUCTION TO CREATING VISUALIZATIONS WITH MATPLOTLIB
#https://app.datacamp.com/learn/courses/introduction-to-data-visualization-with-matplotlib
#Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
#It provides an object-oriented API for embedding plots into applications that use general-purpose GUI toolkits, such as Tkinter, wxPython, Qt, or GTK.

#import matplotlib.pyplot as plt --> to import the matplotlib library
#fig, ax = plt.subplots() --> to create a figure and axis object
#ax.plot(x, y) --> to plot x and y data on the axis
#plt.show() --> to display the plot

#!!!Customizing Plots with Markers, Line Styles, and Colors!!!
#!!!Markers and Line Styles!!!
#ax.plot(x, y, marker='o', linestyle='--') --> to plot with circle markers and dashed line style
#ax.plot(x, y, marker='s', linestyle=':') --> to plot with square markers and dotted line style
#ax.plot(x, y, marker='^', linestyle='-.') --> to plot with triangle markers and dash-dot line style
#ax.plot(x, y, marker='x', linestyle='-') --> to plot with x markers and solid line style   
#You can customize the markers and line styles in your plots using the marker and linestyle parameters in the plot() function.
#https://matplotlib.org/stable/api/markers_api.html --> For more marker styles
#https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html --> For more line styles
#eliminate markers or line styles by setting the respective parameter to 'None': marker=None or linestyle=None
#colors can be customized using the color parameter: color='red' or color='#FF5733' (hex code)
#You can customize the colors in your plots using the color parameter in the plot() function.
#https://matplotlib.org/stable/gallery/color/named_colors.html --> For more color options
#Multiple customizations can be combined in a single plot: ax.plot(x, y, marker='o', linestyle='--', color='blue')  --> to plot with circle markers, dashed line style, and blue color  
#You can combine multiple customizations in a single plot using the marker, linestyle, and color parameters in the plot() function.
#!!!Adding Titles and Labels!!!
#ax.set_title("Plot Title") --> to set the title of the plot
#ax.set_xlabel("X-axis Label") --> to set the label of the x-axis
#ax.set_ylabel("Y-axis Label") --> to set the label of the y-axis
#You can add titles and labels to your plots using the set_title(), set_xlabel(), and set_ylabel() methods of the axis object.
#!!!Adding Legends!!!
#ax.plot(x1, y1, label='Data 1') --> to plot the first dataset with a label
#ax.plot(x2, y2, label='Data 2') --> to plot the second dataset with a label
#ax.legend() --> to display the legend
#You can add legends to your plots using the label parameter in the plot() function and the legend() method of the axis object.
#!!!Customizing Axes Limits and Ticks!!!
#ax.set_xlim(0, 10) --> to set the x-axis limits from 0 to 10
#ax.set_ylim(0, 100) --> to set the y-axis limits from 0 to 100
#ax.set_xticks([0, 2, 4, 6, 8, 10]) --> to set custom x-axis ticks
#ax.set_yticks([0, 20, 40, 60, 80, 100]) --> to set custom y-axis ticks
#You can customize the axes limits and ticks in your plots using the set_xlim(), set_ylim(), set_xticks(), and set_yticks() methods of the axis object.
#!!!Saving Plots to Files!!!
#plt.savefig("plot.png") --> to save the plot as a PNG file
#plt.savefig("plot.pdf") --> to save the plot as a PDF file
#You can save your plots to files using the savefig() function from the matplotlib.pyplot module.

#!!!Creating small multiples (subplots)!!!
#fig, axs = plt.subplots(2, 2) --> to create a 2x2 grid of subplots
#axs[0, 0].plot(x1, y1) --> to plot on the first subplot
#axs[0, 1].plot(x2, y2) --> to plot on the second subplot
#axs[1, 0].plot(x3, y3) --> to plot on the third subplot
#axs[1, 1].plot(x4, y4) --> to plot on the fourth subplot
#to make sure that all have the same range of y-axis values: sharey=True in plt.subplots()
#plt.show() --> to display the subplots
#You can create small multiples (subplots) using the subplots() function from the matplotlib.pyplot module and accessing each subplot using indexing.

#THE RELEVANCE OF CONFIDENCE INTERVALS IN DATA ANALYSIS
#https://www.youtube.com/watch?v=tFWsuO9f74o
#A confidence interval is a range of values that is likely to contain a population parameter with a certain level of confidence.
#It is used to estimate the uncertainty around a sample statistic, such as the mean or proportion.
#Common confidence levels are 90%, 95%, and 99%.
#A 95% confidence interval means that if we were to take many samples and calculate the confidence interval for each sample, approximately 95% of those intervals would contain the true population parameter.
#The width of the confidence interval depends on the sample size, variability in the data, and the chosen confidence level.
#Larger sample sizes and lower variability lead to narrower confidence intervals, while higher confidence levels result in wider intervals.
#Confidence intervals are useful for making inferences about a population based on a sample and for comparing different groups or treatments in experiments.



```
